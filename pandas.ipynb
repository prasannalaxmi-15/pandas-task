```python
import numpy as np
import pandas as pd
import os
```

```python
df = pd.read_csv(r"C:\Users\prasa\Downloads\archive (2)\Dataset.v2\Mix.csv")  
```

```python
df
```

```python
df["Heart Sound Type"] = df["Heart Sound Type"].astype(str).str.strip().str.title()
save_path = r"C:\Users\prasa\Desktop\pandas\Heart_Sound_Types"
os.makedirs(save_path, exist_ok=True)
print("Available Heart Sound Types:\n", df["Heart Sound Type"].unique(), "\n")
for sound_type, data in df.groupby("Heart Sound Type"):
    filename = sound_type.replace(" ", "_") + ".csv"
    full_path = os.path.join(save_path, filename)
    data.to_csv(full_path, index=False)
```
